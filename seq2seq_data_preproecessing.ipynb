```python
import pandas as pd
```

```python
df = pd.read_csv("카페_train.csv")
```

```python
df
```

```python
#모든 줄의 "발화자","발화문","카테고리","QA번호","QA여부","인텐트" 컬럼 리턴
cafe_df = df.loc[: ,["발화자","발화문","카테고리","QA번호","QA여부","인텐트"]]
cafe_df
```

```python
#키오스크는 특징상 고객의 물음의 대답하는 형식

#발화자가 c (고객) 이면서 QA여부가 q인 행 리턴
cafe_df[( cafe_df["발화자"]=="c") & ( cafe_df["QA여부"]=="q") ]
```

```python
#발화자가 c (고객) 이면서 QA여부가 q인, QA여부가 5인 행 리턴
#결과값이 보이다싶이 특정번호가 중복으로 나오는 잘못된 데이터가 많음
cafe_df[( cafe_df["발화자"]=="c") & ( cafe_df["QA여부"]=="q") & ( cafe_df["QA번호"]==5) ]
```

```python
# 인텐트에 반품, 멤버쉽, 예약 등의 키오스크에는 필요하지 않은 데이터 제거 

#발화자가 c (고객) 이면서 QA여부가 q
# "인텐트" 에  결제 주문 포장 제품 이 포함된 행
cq_df = cafe_df[
         ( cafe_df["발화자"]=="c") 
         & ( cafe_df["QA여부"]=="q") 
         & ( cafe_df["인텐트"].str.contains("결제|주문|포장|제품"))
             ]
cq_df
```

```python
#위에서 봤던거처럼 qa번호가 연달아서 나오지 않는경우 비정상 데이터들이 들어가있음
#연달아서 데이터가 들어가있지 않는 경우 제거하기위해서 index번호를 통해 처리

#고객 질의 데이터 프레임의 인덱스를 리턴
cq_df.index
```

```python
#고객 질의 데이터 프레임의 인덱스 + 1를 리턴 (직원 응답)
cq_df.index+1
```

```python
##고객 질의 데이터 프레임의 인덱스 + 1 행  (직원 응답)

ca_df = cafe_df.loc[ cq_df.index+1 , : ]
ca_df
```

```python
#발화자가 s가 아닌경우 고객에 질문에 대한 답변이 아님

#발화자가 s (직원) 인 행 조회
ca_df[ ca_df ["발화자"]=="s" ]
```

```python
#발화자가 s (직원) 인 행 조회의 인덱스 조회
ca_df[ ca_df ["발화자"]=="s" ].index
```

```python
#발화자가 s (직원) 인 행 조회의 인덱스-1 (윗줄)  인덱스 조회 => 고객 질의
ca_df[ ca_df ["발화자"]=="s" ].index - 1
```

```python
#발화자가 s (직원) 인 행 조회의 인덱스-1 (윗줄)  인덱스 조회 => 고객 질의
ca_index = ca_df[ ca_df ["발화자"]=="s" ].index - 1
ca_index
```

```python
#발화자가 s (직원) 인 행 조회의 인덱스-1 (윗줄)  조회 => 고객 질의
cq_df2 = cq_df.loc [ ca_index]
#발화자가 고객이고 질문하는 데이터
cq_df2
```

```python
#발화자가 s (직원) 인 행 조회
ca_df2 = ca_df[ ca_df ["발화자"]=="s" ]
ca_df2
```

```python
#drop=True: 인덱스 삭제
# inplace=True : 새로운 인덱스 적용
ca_df2.reset_index(drop=True,inplace=True)
cq_df2.reset_index(drop=True,inplace=True)
```

```python
#직원 응답
ca_df2
```

```python
cq_df2
```

```python
#고객 질의 발화문 열 조회
cq_df2[["발화문"] ]
```

```python
#seq2seq모델의 전처리는 일반적으로 입력시퀀스(질문)과 출력시퀀스(답변)에 대한 데이터가 한 행에 있어야함

#고객 질의 발화문 열 과 직원 응답 발화문 열 병합
#axis=0이면 열로 이어붙이고 1이면 행에 붙힘
cafe_data = pd.concat([ cq_df2[["발화문"] ] , ca_df2[["발화문"] ] ] , axis=1)

#컬럼 명을  q a 로 변환
cafe_data.columns = ['q','a']

cafe_data
```

```python
#cafe.csv로 저장
cafe_data.to_csv (
                   "cafe_train.csv",
                    index=False,
                   encoding="utf-8-sig"
                 )
```
